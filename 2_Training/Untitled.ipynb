```python
import torchvision
import itertools
import matplotlib.pyplot as plt
```

```python
filename = '/notebooks/data/Datasets/CPTAD/Videos/WUDQ6wSbMOw01.mp4'

video = torchvision.io.VideoReader(filename, "video")
metadata = video.get_metadata()
print(metadata)
```

```python
fps = metadata["video"]['fps'][0]
frames = []

start_s = 200/fps
for frame in itertools.islice(video.seek(start_s), 16):
    frames.append(frame['data'])
```

```python
plt.figure(figsize=(12, 12))
for i in range(16):
    plt.subplot(4, 4, i + 1)
    plt.imshow(frames[i].permute(1, 2, 0))
    plt.axis("off")
```
